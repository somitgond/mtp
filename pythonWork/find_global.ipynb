```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data_naqm = pd.read_csv("results-naqm.csv")
data_aqm = pd.read_csv("results-aqm.csv")
data_naqm_10 = data_naqm[:10]
data_aqm_10 = data_aqm[:10]
data_naqm_11 = data_naqm[10:]
data_aqm_11 = data_aqm[10:]
```

```python
data_naqm
```

```python
data_aqm
```

```python
width = 0.25
x = np.arange(10)
multiplier = 0
fig, ax = plt.subplots(layout='constrained')
label = list(data_naqm_10["Random Seed"])
datasets = {
    "Q: 2084": list(data_naqm_10["Global Sync Value"]), 
    "Q: 25": list(data_aqm_10["Global Sync Value"])
    }
for l, d in datasets.items():
    offset = multiplier*width
    rects = ax.bar(x+offset, d, width, label = l)
    # ax.bar_label(rects, padding=3)
    multiplier += 1
# plt.bar(list(data_naqm_10["Simulation_number"]), list(data_naqm_10["Global Sync Value"]), width = 0.8, label="no aqm")
# plt.bar(list(data_aqm_10["Simulation_number"]), list(data_aqm_10["Global Sync Value"]), width = 0.8, label = "aqm")
plt.xlabel("Random Seed")
plt.ylabel("Global Sync Value")
plt.title("Global Sync Value vsRTT")
ax.set_xticks(x+width, label)
# plt.grid()
plt.legend(loc='lower left')
plt.savefig
```

```python
plt.scatter(list(data_naqm_11["RTT"]), list(data_naqm_11["Global Sync Value"]), label="no aqm")
plt.scatter(list(data_aqm_11["RTT"]), list(data_aqm_11["Global Sync Value"]), label = "aqm")
plt.xlabel("RTT")
plt.ylabel("Global Sync Value")
plt.title("Global Sync Value vs RTT")
plt.grid()
plt.legend()
plt.savefig
```

```python
plt.scatter(list(data_naqm_11["RTT"]), list(data_naqm_11["Average Throughput"]), label="no aqm")
plt.scatter(list(data_aqm_11["RTT"]), list(data_aqm_11["Average Throughput"]), label = "aqm")
plt.xlabel("RTT")
plt.ylabel("Average throughput")
plt.title("Average throughput vs RTT")
plt.grid()
plt.legend()
```

```python
plt.scatter(list(data_naqm_11["RTT"]), list(data_naqm_11["Effective Delay"]), label="no aqm")
plt.scatter(list(data_aqm_11["RTT"]), list(data_aqm_11["Effective Delay"]), label = "aqm")
plt.xlabel("RTT")
plt.ylabel("Effective Delay")
plt.title("Effective Delay vs RTT")
plt.grid()
plt.legend()
```

```python
import numpy as np

# Load data
data = np.loadtxt('qsizeTrace-dumbbell_q30.txt', usecols=1)

# Compute autocorrelation with lag 50
lag = 70
mean = np.mean(data)
var = np.var(data)
autocorr1 = np.correlate(data - mean, data - mean, mode='full') / (len(data) * var)
result = autocorr1[len(data)-1 + lag]
print(result)
```

```python
plt.plot(autocorr1)
plt.title("Autocorrelation q=25")
```
